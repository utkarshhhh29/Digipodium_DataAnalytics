```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,10),layout='constrained')
axs[0,0].hist('sepal_width', data=iris, bins=20)
axs[0,1].hist('sepal_length', data=iris, bins=20)
axs[1,0].hist('petal_width', data=iris, bins=20)
axs[1,1].hist('petal_length', data=iris, bins=20)
axs[0,0].set_title('sepal_width')
axs[0,1].set_title('sepal_length')
axs[1,0].set_title('petal_width')
axs[1,1].set_title('petal_length')
plt.show()
```

```python
plt.scatter(x='sepal_width', y='sepal_length', data=iris, color = 'green')
plt.show()
```

```python
def convert_to_number(name):
    if name == 'setosa':
        return 1
    if name == 'virginica':
        return 2
    return 3

iris['spc'] = iris['species'].apply(convert_to_number)
iris
```

```python
plt.scatter(x='sepal_width', y='sepal_length', data=iris, c='spc', cmap='rainbow')
plt.colorbar()
plt.show()
```
