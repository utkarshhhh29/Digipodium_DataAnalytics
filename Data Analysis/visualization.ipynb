```python
x = np.arange(-5,5,.2)
y1 = np.sin(x)
y2 = np.cos(x)
fig, axs = plt.subplots(nrows=2,ncols=1,figsize=(10,8)) #axs is arrayy object
axs[0].plot(x,y1,'r--')
axs[1].plot(x,y2,'b--')
axs[0].set_title('sin(x)')
axs[1].set_title('cos(x)')
plt.show()
```

implicit vs explicit plot

```python
#implicit
x = np.linspace(-10,10,25)
plt.plot(x,x,'-',label='linear')
plt.plot(x,x**2,'--',label='quadratic')
plt.plot(x,x**3,'o:',label='cubic')
plt.xlabel('my x label')
plt.ylabel('my y label')
plt.title('Simple graph with implicit graph setting')
plt.legend()
plt.show()         
```

```python
#explicit
x = np.linspace(-10,10,25)
fig, ax = plt.subplots()
ax.plot(x,x,'g-',label='linear')
ax.plot(x,x**2,'r--',label='quadratic')
ax.plot(x,x**3,'o:',label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
plt.savefig('Plot1.png',bbox_inches='tight',dpi=300) #save the figure to file
plt.show() 

```

```python
#using dictionary for graph
data = {
    "a": np.arange(50),
    "b": np.random.randint(0,50,50),
    "c": np.random.randn(50),
    "d": np.arange(50) + 10 * np.random.randn(50) 
}
data['d']= np.abs(data['d'])*10

fig, ax = plt.subplots(figsize=(10,3))
ax.scatter('a','b',c='c',s='d',data=data,cmap='jet')
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
ax.set_title('Scatter Plot')
plt.show()
```

```python
data = {
    "Name": ['Jett','Raze','Breach','Cypher','Omen'],
    "English": [43,41,65,53,60],
    "Maths": [22,53,24,62,40],
    "Computer": [43,25,56,43,42]
}
```

```python
def my_custom_plotter(ax,x_value,y_value,param_dict):
    out = ax.bar(x_value,y_value,**param_dict)
    ax.set_title('Visualising Marks')
    ax.set_xlabel('Marks')
    ax.set_ylabel('Counts')
    return out
```

```python
fig, axs = plt.subplots(1,3,figsize=(15,3))
my_custom_plotter(axs[0],'Name','English',{'data':data,'color':'red','alpha':.3}) #alpha is for transparency of color
my_custom_plotter(axs[1],'Name','Maths',{'data':data,'color':'green'})
my_custom_plotter(axs[2],'Name','Computer',{'data':data,'color':'blue'})
plt.show()
```
