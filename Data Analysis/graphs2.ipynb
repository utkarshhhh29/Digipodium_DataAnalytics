```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #basic graphs
import seaborn as sns #advanced graphs
import plotly.express as px #interactive graphs
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

# Categorical Data
    
- columns that have text are generally categorical
- the first step is to count the number of unique values in column
- then plot the count of each category
- following charts are a good way to visualize categorical data
    - bar chart
    - pie chart
    - donut chart
    - line chart
    - area chart

```python
print(titanic['survived'].unique()) #to check if the column is ccategorical or not
print(titanic['pclass'].unique())
```

```python
survival_count = titanic['survived'].value_counts()
survival_count.plot(kind='pie',autopct='%.1f%%', title='Survival Rates') #autopct = auto percentage
plt.show()
```

```python
floors = titanic['pclass'].value_counts()
floors
```

```python
survival_count.plot(
    kind='pie',
    autopct='%.1f%%',
    title='Survival Rate',
    labels=['Survived','Not Survived'],
    colors=['#66b3ff','#ff9999'],
    labeldistance=1.1,
    wedgeprops={'width':0.5}
    )
plt.show()
```

```python
floors = titanic['pclass'].value_counts()
floors.plot(kind='pie',autopct='%.1f%%', title='Floors')
plt.show()
```

```python
floors.plot(
    kind='pie',
    autopct='%.1f%%',
    title='Floors',
    labels=['1st floor','2nd floor','3rd floor'],
    colors=['#66b3ff','#ff9999','orange'],
    labeldistance=1.1,
    wedgeprops={'width':0.7}
    )
plt.show()
```

```python
gender = titanic['sex'].value_counts()
gender.plot(kind='pie',autopct='%.1f%%', title='Sex')
plt.show()
```

```python
gender.plot(
    kind='pie',
    autopct='%.1f%%',
    title='Sex',
    labels=['Male','Female'],
    colors=['Brown','Pink'],
    labeldistance=1.1,
    wedgeprops={'width':0.7}
    )
plt.show()
```

```python
clas = titanic['class'].value_counts()
clas.plot(kind='pie',autopct='%.1f%%', title='Class')
plt.show()
```

```python
clas.plot(
    kind='pie',
    autopct='%.1f%%',
    title='Class',
    labels=['First','Second','Third'],
    colors=['Yellow','Blue','green'],
    labeldistance=1.1,
    wedgeprops={'width':0.7}
    )
plt.show()
```
